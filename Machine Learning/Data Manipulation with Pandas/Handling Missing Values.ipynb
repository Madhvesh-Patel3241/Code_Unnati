# Handling Missing Values

## Checking Missing values using isnull() & notnull()

```python
#importing panadas as pd
import pandas as pd 

#importing numpy as np
import numpy as np

# Dictionary of lists
dict = {
    'First Score':[100, 90, np.nan, 95],
    'Second Score':[30, 45, 56, np.nan],
    'Third Score':[50, 40, 80, 98]
       }

#creating a dataframe from list
df = pd.DataFrame(dict)


# using isnull() function
df.isnull()

#print the Dataframe
df
```

## Dropping missing value using Dropna()

```python
#importing panadas as pd
import pandas as pd 

#importing numpy as np
import numpy as np

# Dictionary of lists
dict = {
    'First Score':[100, 90, np.nan, 95],
    'Second Score':[30, np.nan, 45, 56],
    'Third Score':[52, 40, 80, 98],
    'Fourth Score':[86, np.nan, np.nan, 65]
       }

#creating a dataframe from list
df = pd.DataFrame(dict)


# using dropna() function
df.dropna()
```
