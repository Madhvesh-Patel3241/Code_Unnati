```python
# ticks are the markers denoting data points on axes
#importing libraries
import matplotlib.pyplot as plt
import numpy as np

# x axis values
x = [5,10,15,0,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]


plt.plot(x,y, 'b')
plt.xlabel("x - axis")
plt.ylabel("y - axis")
#giving a title to my graph
plt.title("My First Graph")

#here we set the size for ticks, rotation and color value

plt.tick_params(axis="x", labelsize=18, labelrotation=-60, labelcolor='r')
plt.tick_params(axis="y", labelsize=18, labelrotation=30, labelcolor='g')

#function to show the plot 
plt.show()
```

```python
#IMPORT MATPLOTLIB.PYPLOT
import matplotlib.pyplot as plt
import numpy as np

#DRAW PLOT AS FIG
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')
#TEXT ANNOTATION TO THE PLOT WHERE IT INDICATE MAXIMUM VALUE OF THE CURV
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4), arrowprops=dict(facecolor='blue', shrink=0.05))
#TEXT ANNOTATION TO THE PLOT WHERE IT INDICATE MINIMUM VALUE OF THE CURV
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6), 
            arrowprops=dict(arrowstyle="wedge",
                            connectionstyle="angle3,angleA=0,angleB=-90"));
```

```python
import matplotlib.pyplot as plt
w = 4
h = 3
d = 70
plt.figure(figsize=(w, h), dpi=d)
plt.axis([0, 5, 0, 5])
x = [0, 3, 5]
y = [1, 4, 3.5]
label_x = 1
label_y = 4
arrow_x = 3
arrow_y = 4

arrow_properties = dict(
    facecolor="black", width=0.5,
    headwidth=4, shrink=0.1)

plt.annotate(
    "maximum", xy=(arrow_x, arrow_y),
    xytext=(label_x, label_y),
    arrowprops=arrow_properties)
plt.plot(x, y)
```

```python
# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# defining all 3 axes
z = np.linspace(0, 1, 100)
x = z * np.sin(30 * z)
y = z * np.cos(30 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```
